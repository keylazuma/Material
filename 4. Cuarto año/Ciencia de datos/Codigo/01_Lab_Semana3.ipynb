#### Creando variables
Una variable TensorFlow es la forma recomendada de representar el estado compartido y persistente que manipula un programa.

Para crear una variable, se debe proporcionar un valor inicial. La variable tf.Variable tendrá el mismo tipo de datos que el valor de inicialización. Una variable se ve y actúa como un tensor y, de hecho, es una estructura de datos respaldada por un tensor tf.Tensor. Al igual que los tensores, tienen un tipo de datos y una forma, y ​​se pueden exportar a NumPy.

Para un mejor rendimiento, TensorFlow intentará colocar tensores y variables en el dispositivo más rápido compatible con su tipo de datos. Esto significa que la mayoría de las variables se colocan en una GPU si hay una disponible.

Las variables son como esas partes ajustables, diseñadas para cambiar según sea necesario. Son las herramientas especiales que se utilizan para ajustar el modelo durante el entrenamiento, lo que lo ayuda a encontrar sus mejores configuraciones.

Dos diferencias claves entre tensores y variables:
1. Los tensores son fijos; las variables son dinámicas y se pueden actualizar
2. Los tensores representan datos de entrada/salida y resultados intermedios; las variables contienen y actualizan los parámetros del modelo.

### Gradientes y diferenciación automática

La diferenciación automática es útil para implementar algoritmos de aprendizaje automático como la retropropagación para entrenar redes neuronales.

Para diferenciar automáticamente, TensorFlow necesita recordar qué operaciones se realizan en qué orden durante el paso hacia adelante. Luego, durante el paso hacia atrás, TensorFlow recorre esta lista de operaciones en orden inverso para calcular los gradientes.

**Usando `GradientTape`**

TensorFlow proporciona la API tf.GradientTape para la diferenciación automática; es decir, el cálculo del gradiente de un cálculo con respecto a algunas entradas, normalmente tf.Variables. TensorFlow "registra" las operaciones relevantes ejecutadas dentro del contexto de un tf.GradientTape en una "cinta". TensorFlow luego utiliza esa cinta para calcular los gradientes de un cálculo "registrado" utilizando la diferenciación en modo inverso.

```python
x = tf.Variable(3.0)

with tf.GradientTape() as tape:
  y = x**2
```

Cuál es la derivada de x al cuadrado?

Una vez que haya registrado algunas operaciones, use GradientTape.gradient(target, fuentes) para calcular el gradiente de algún objetivo (a menudo una pérdida) en relación con alguna fuente (a menudo las variables del modelo):

```python
dy_dx = tape.gradient(y, x) # Gradiente de y con respecto a x
print(dy_dx.numpy())
```

Otro ejemplo

Asumiento solamente la parte lineal de la operación de las neuronas (wTx + b) de una capa de entrada de 3 neuronas a una capa de salida de 1 neurona

```python
# Asumiento b + wx: ecuación de la regresión lineal o base de una red neuronal sin la activación)

w = tf.Variable(tf.random.normal((3, 1)), name='w')
b = tf.Variable(tf.zeros((1,1), dtype=tf.float32), name='b')
x = tf.reshape(tf.constant([1., 2., 3.]),(3, 1))
y_true = tf.Variable(tf.ones((1,1), dtype=tf.float32), name='b')

print("w:", w, "\n b:", b, "\n x:", x)

with tf.GradientTape(persistent=True) as tape:
  y_pred = tf.matmul(tf.transpose(w), x)  + b
  loss = ((y_pred - y_true)**2)/2 # Obtiene la media del tensor

print(y_pred, loss)
```

Para obtener el gradiente de pérdida con respecto a ambas variables, puede pasar ambas como fuentes al método de gradiente. La cinta es flexible en cuanto a cómo se pasan las fuentes y aceptará cualquier combinación anidada de listas o diccionarios y devolverá el gradiente estructurado de la misma manera (consulte tf.nest).

```python
[dl_dw, dl_db] = tape.gradient(loss, [w, b])
print(dl_dw, dl_dw.shape)
print("---")
print(dl_db, dl_db.shape)
```

```python
# Derivada con respecto a la función de pérdida (ypred - ytrue) * x
# Ejemplo w1:
i = 0
deriv_loss_pred = (y_pred - y_true)
print(x[i][0]*deriv_loss_pred)

# Ejemplo w3
i = 2
print(x[i][0]*deriv_loss_pred)
```

### Ejemplo de clasificador con TensorFlow

**Generando datos de dos clases con valores aleatorios de 2 variables**

```python
n_clase = 1000
muestras_0 = np.random.multivariate_normal(
    mean=[0, 3],
    cov=[[1, 0.5],[0.5, 1]],
    size=n_clase)
muestras_1 = np.random.multivariate_normal(
    mean=[3, 0],
    cov=[[1, 0.5],[0.5, 1]],
    size=n_clase)

print(muestras_0.shape, muestras_0)
print(muestras_1.shape, muestras_1)
```

**Uniendo los datos de entrada**

```python
inputs = np.vstack((muestras_0, muestras_1)).astype(np.float32)
inputs = tf.convert_to_tensor(inputs, dtype=tf.float32)
inputs.shape
```

```python
inputs[0] # observación 1
```

**Generación de variable dependiente de clasificación binaria**

```python
targets = np.vstack((np.zeros((n_clase, 1), dtype="float32"),
                     np.ones((n_clase, 1), dtype="float32")))
targets = tf.convert_to_tensor(targets, dtype=tf.float32)

targets.shape
```

```python
targets[0]
```

**Graficando**

```python
import matplotlib.pyplot as plt
plt.scatter(inputs[:, 0], inputs[:, 1], c=targets[:, 0])
plt.xlabel("Variable 1")
plt.ylabel("Variable 2")
plt.title("Datos creados para clasificación lineal con tensorflow")
plt.show()
```

**Creando las variables de la ecuación de clasificación lineal**

W = pesos
b = intercepto

```python
input_dim = 2
output_dim = 1
# Iniciación aleatoria de pesos
W = tf.Variable(initial_value=tf.random.uniform(shape=(input_dim, output_dim)))
b = tf.Variable(initial_value=tf.zeros(shape=(output_dim,)))
print(W, b)
```

**Forward pass/ Propagación hacia delante**

```python
def model(inputs):
    return tf.matmul(inputs, W) + b

model(inputs) # Resultado de operación
```

¿Qué le falta a la ecuación para ser una regresión logística?

```python
def model(inputs):
    linear_output = tf.matmul(inputs, W) + b
    # Aplicando sigmoirde
    logistic_output = tf.sigmoid(linear_output)
    return logistic_output

model(inputs) # Resultado de operación
```

**Función de pérdida: entropía cruzada binaria**

```python
def bin_loss_(targets, predictions):
  loss = -tf.reduce_mean(targets * tf.math.log(predictions) + (1 - targets) * tf.math.log(1 - predictions))
  return loss

loss_res = bin_loss_(targets, model(inputs))
print(loss_res)
```

**Función de entrenamiento: cálculo de gradiente**

```python
def training_step(inputs, targets, learning_rate = 0.1):
  with tf.GradientTape() as tape:
    # Forward pass
    predictions = model(inputs)
    # Función de pérdida
    loss = bin_loss_(targets, predictions)
  # Obtención de gradiente
  grad_loss_wrt_W, grad_loss_wrt_b = tape.gradient(loss, [W, b])
  # Actualización de los parámetros considerando el tasa de aprendizaje
  W.assign_sub(grad_loss_wrt_W * learning_rate) # tasa de aprendizaje
  b.assign_sub(grad_loss_wrt_b * learning_rate)
  return loss
```

**Batch entrenamiento**

```python
# Entrenando por 15 veces
for step in range(15):
    loss = training_step(inputs, targets)
    print(f"Loss at step {step}: {loss:.4f}")
```

```python
# Comparación de predicciones vs valores reales
predictions = model(inputs)

fig, ax = plt.subplots(1,2, figsize=(10,5))
ax[0].scatter(inputs[:, 0], inputs[:, 1], c=predictions[:, 0] > 0.5)
ax[0].set_title("Predicciones")
ax[1].scatter(inputs[:, 0], inputs[:, 1], c=targets[:, 0])
ax[1].set_title("Valores reales")
plt.show()
```

## Redes neuronales con Keras

Cuando entrenamos redes neuronales con keras, ya hay funciones pre-existentes que representan capas, funciones de activación, funciones de pérdida, etc. A continuación veremos los componentes de mayor uso de keras para entrenar una red neuronal

Una capa `Dense`: En las redes neuronales, las capas son elementos fundamentales. En particular, una capa densa (también conocida como capa completamente conectada) es una capa en la que cada neurona recibe información de todas las neuronas de la capa anterior.

Output=Activation(Input⋅W+b)

```python
from tensorflow.keras import layers

layer = layers.Dense(32, activation="relu") # Capa densa de 32 unidades y activación relu

# Imprimir detalles de la configuración:
print(layer.get_config())
```

```python
# Ejemplo con 10 variables, una observación
import numpy as np
dummy_input = np.random.random((1, 10))
print(dummy_input)
# Pasamos los datos en esta capa
output  = layer(dummy_input)

# Pesos e intercepto
weights, biases = layer.get_weights()
print(f"Pesos de tamaño {weights.shape}, un ejemplo:", weights[0])
print(f"Sesgos/interceptos de tamaño {biases.shape}:", biases)
print(f"Resultado de tamaño {output.shape}:", output)
```

Añadiendo más capas:

```python
from tensorflow.keras import layers, Sequential
# Construcción secuencial donde añadimos 4 capas
input_dim = 10 # Número de variables
num_clases = 5
model = Sequential([
    layers.Dense(32, activation="relu",input_shape=(input_dim,)), # capa de entrada + primera capa oculta
    layers.Dense(64, activation="relu"), # segunda capa oculta
    layers.Dense(32, activation="relu"), # tercera capa oculta
    layers.Dense(num_clases, activation="softmax") # capa de salida
])

# Configuracion
config = model.get_config()

# Función para imprimir de manera estructurada
def print_layer_config(layer_config):
    print(f"Capa: {layer_config['class_name']}")
    for key, value in layer_config['config'].items():
        print(f"  {key}: {value}")
    print()

# Print the model configuration
print(f"Nombre del modelo: {config['name']}\n")
for layer in config['layers']:
    print_layer_config(layer)
```
